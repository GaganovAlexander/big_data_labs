#### Подзадание 2
Для начала создаю столбец в котором будет храниться только год-месяц поездки

```python
df['year-month'] = df.pickup_datetime.dt.strftime('%Y-%M')
```

Потом провожу необходимый рассчёт

```python
df1 = df[['year-month', 'id']].groupby('year-month').count() # Я выбрал id в качестве столбца, по которому будет производиться count так как он однозначно показывает отдельную поездку
df1.columns = ['count']
df1
```
